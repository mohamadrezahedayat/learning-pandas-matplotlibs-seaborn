### exploring groups

```python
titanic = pd.read_csv("DataAnalysis/data/titanic.csv")
titanic['age']=titanic['age'].replace(['?'],[None]).astype('float')
# or titanic['age']=pd.to_numeric(titanic.age,errors='coerce')
```

```python
df = titanic[['pclass','survived','sex','age']]
```

```python
gbo = df.groupby('sex')
```

### agg with custom functions


```python
def range(x):
  return x.max() - x.min()

titanic.groupby('pclass')['age'].agg(range)
```

```python
def range(x):
  return x.max() - x.min()

titanic.groupby('pclass')['age'].agg(['min','max',range])
```

```python
def count_nulls(s):
  return s.size - s.count()

titanic.groupby('pclass').agg(count_nulls)
```

### named aggregation

```python
carstocks.groupby('Symbol').agg({'Open':['min','max'],'Close':['min','max']})
```

```python
carstocks.groupby('Symbol').agg(
  min_Open=('Open','min'),
  max_Open=('Open','max'),
  min_Close=('Close','min'),
  max_Close=('Close','max'),
)
```
