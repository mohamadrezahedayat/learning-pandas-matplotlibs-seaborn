### combining conditions using AND


```python
woman = titanic.sex == 'female'
died = titanic.survived == 0
```

### combining conditions using OR

```python
houses[(houses.yr_built>2010 )| (houses.yr_renovated>2010) ]
```

### bitwise negation

```python
df = titanic.head()
women = df.sex =='female'
df[~women]
```

```python
newly_built = houses.yr_built>2013
newly_renovated = houses.yr_renovated>2013
recent_homes = newly_built | newly_renovated
houses[ ~recent_homes]
```
