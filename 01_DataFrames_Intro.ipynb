# 5 - overriding headers
### - for example we want to turn header names to lower case

```python
dt= pd.read_csv("DataAnalysis/data/nst-est2020.csv")
names= dt.columns
lower_names=[x.lower() for x in names]
lower_names
```

```python
dt_with_new_header = pd.read_csv("DataAnalysis/data/nst-est2020.csv",names=lower_names,header=0)
dt_with_new_header.head(5)
```
