## Bar plots

```python
planets = ['Spinach','Turnip','Rhubarb','Broccoli','Kale']
died = [10,25,5,30,21]
germinated = [74, 88,56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(planets,germinated, width=0.5)
plt.title('Germination Rates')
plt.ylabel('Number of Seedlings')
```

```python
plt.figure(figsize=(6,6))
plt.bar(planets,germinated)
plt.bar(planets,died)
```

```python
plt.figure(figsize=(6,6))
plt.bar(planets,germinated, label='Germinated Seedlings')
plt.bar(planets,died , bottom=germinated, label='Failed Seedlings')
plt.legend(shadow=True,frameon=True, facecolor='white')
```

#### horizontal bar

```python
plt.barh(planets,germinated, label='Germinated Seedlings')
```

```python
import pandas as pd
titanic = pd.read_csv("DataAnalysis/data/titanic.csv")
titanic.age = pd.to_numeric(titanic.age,errors='coerce')
```

```python
plt.hist(titanic.age)
```

```python
first_class = titanic[titanic.pclass ==1]['age']
second_class = titanic[titanic.pclass ==2]['age']
third_class = titanic[titanic.pclass ==3]['age']
plt.hist(first_class, label = '1st Class',alpha=.7, bins = 20)
plt.hist(second_class, label = '2nd Class',alpha=.7, bins = 20)
plt.hist(third_class, label = '3rd Class',alpha=.7, bins = 20)
plt.legend()
```

### Subplots

```python
nums = np.arange(5)
plt.figure(figsize=(10,4))
plt.plot(nums)
```

```python
nums = np.arange(5)
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.plot(nums,color='purple')
plt.title('X')
plt.subplot(1,3,2)
plt.plot(nums*nums,color='orchid')
plt.title('XX')
plt.subplot(1,3,3)
plt.plot(nums**3, color='olive')
plt.title('XXX')
```

```python
nums = np.arange(5)
plt.figure(figsize=(10,4))
plt.suptitle('our first subplot')

plt.subplot(1,3,1)
plt.plot(nums,color='purple')
plt.title('X')

plt.subplot(1,3,2)
plt.plot(nums*nums,color='orchid')
plt.title('XX')

plt.subplot(1,3,3)
plt.plot(nums**3, color='olive')
plt.title('XXX')

plt.tight_layout()
```
